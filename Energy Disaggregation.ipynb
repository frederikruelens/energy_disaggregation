# Energy disaggregations dataset - http://redd.csail.mit.edu/

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%matplotlib notebook
```

# Labels of the measurements

```python
# measuremnts 
labels = open("house_1/labels.dat").read()
print labels
n_channels = labels.count('\n')
```

# Import power measurements

```python
time_stamp_channels     = [] 
power_readings_channels = [] 

for i_channel in range(0,n_channels):
    print "channel "  + str(i_channel)
    file_name      = "house_1/channel_" + str(i_channel+1) +".dat"
    time_stamp     = []
    power_readings = []
    for line in open(file_name):
        columns = line.split(" ") #place counter for line,seperator as space [" "]
        time_stamp.append(np.int(columns[0].rstrip()))
        power_readings.append(np.float(columns[1].rstrip()))
    time_stamp_channels.append(time_stamp)
    power_readings_channels.append(power_readings)
```

# Visualize aggregated power measurent (one house)

```python
n_steps_one_day    = 60*60*24
time_stamp         = np.asarray(time_stamp)
power_readings     = np.asarray(power_readings)
n_days             = np.shape(power_readings)[0]/n_steps_one_day
aggregated_profile = np.asarray(power_readings_channels[0][0:n_steps_one_day]) + np.asarray(power_readings_channels[1][0:n_steps_one_day])
plt.figure()
plt.plot(aggregated_profile[0:n_steps_one_day], color = "black" , label = 'Aggregated power')
plt.legend()
plt.ylabel("Power [W]")
plt.xlabel("Time [s]")
```

# Visualize power measurement refridgerator

```python
time_stamp_final =  time_stamp_channels[0][n_steps_one_day]
time_stamp_start =  time_stamp_channels[0][0]
time_index       = np.argmin(np.abs(np.asarray(time_stamp_channels[4][:]) - time_stamp_final))

plt.figure()
plt.plot(time_stamp_channels[4][0:time_index],power_readings_channels[4][0:time_index], label = 'Fridge')
plt.plot(time_stamp_channels[0][0:n_steps_one_day], aggregated_profile[0:n_steps_one_day], color = "black" , label = 'Aggregated')
plt.ylabel("Power [W]")
plt.xlabel("Time [s]")
plt.legend()
plt.show()
```
